````python
import numpy as np
import string
import random as rd
alphabet = string.ascii_lowercase

def encrypt_letters(key):
    alphabet_ls = list(alphabet)
    alphabet_encrypted = alphabet_ls[:]
    rd.seed(key)
    rd.shuffle(alphabet_encrypted)
    letter_map = {alphabet_ls[i]: alphabet_encrypted[i] for i in range(26)}
    letter_map_reverse = {alphabet_encrypted[i]: alphabet_ls[i] for i in range(26)}
    return letter_map, letter_map_reverse

def read_file(filename):
    file = open(filename, 'r')
    line_ls = []
    for line in file:
        line_ls.append(line)
    if len(line_ls) <= 10:
        print(*line_ls, sep = '') 
    else:
        for i in range(5):
            print(line_ls[i], end = '')
        print('...')
        for i in range(-5, 0):
            print(line_ls[i], end = '')
    file.close()

def set_key():
    while True:
        try:
            key = int(input('Type the encryption key: '))
            while (key < 100) |  (key > 999): 
                print('Error: it should be a number between 100 and 999.\n')
                key = int(input('Type the encryption key: '))   
            print('')
            break
        except:
            print('Error: it should be a number between 100 and 999.\n')
    return key
````

````python
# set-key
key = set_key()
letter_map, letter_map_reverse = encrypt_letters(key)

while True:
    x = input('>> ')
    
    # set-key 
    if x == 'set-key':
        key = set_key()
        letter_map, letter_map_reverse = encrypt_letters(key)
            
    # bye
    elif x == 'bye':
        print('Bye! Have a nice day.')
        break
        
    # show-key
    elif x == 'show-key':
        print(f'The encryption key is: {key}\n')
        
    # show-message
    elif x == 'show-message':
        try:
            read_file('message.txt')
            print('\n')
        except:
            print("Error. 'message.txt' is not available.\n")
            
    # show-encrypted
    elif x == 'show-encrypted':
        try:
            read_file('encrypted.txt')
            print('\n')
        except:
            print("Error. 'encrypted.txt' is not available.\n")
            
    # show-map
    elif x == 'show-map':
        print('The encryption map is: ')
        for ki, value in letter_map.items():
            print(f'{ki} => {value}', end = '\t')
        print('\nAnd the decryption map is:')
        for ki, value in letter_map.items():
            print(f'{value} => {ki}', end = '\t')
        print('\n')
    
    # encrypt
    elif x == 'encrypt':
        try:
            file = open('message.txt', 'r')
            line_ls = []
            for line in file:
                line_ls.append(line)
            file.close()
            
            file = open('encrypted.txt', 'w')
            for line in line_ls:
                tmp = ''
                for letter in line:
                    if letter in letter_map:
                        tmp = tmp+letter_map[letter]
                    else:
                        tmp = tmp+letter
                file.write(tmp)
            file.close()
            print('Successful encryption.\n')
        except:
            print('Error. The encryption failed.\n')
            
    # decrypt
    elif x == 'decrypt':
        try:
            file = open('encrypted.txt', 'r')
            print('Successful decrypted.')
            print('Your message is:')
            for line in file:
                tmp = ''
                for letter in line:
                    if letter in letter_map_reverse:
                        tmp = tmp+letter_map_reverse[letter]
                    else:
                        tmp = tmp+letter
                print(tmp, end = '')
            file.close()
            print('\n')
            
        except:
            print('Error. The decryption failed.\n')
        
        
    
    else: 
        print('Error. Invalid command.\n')
````
